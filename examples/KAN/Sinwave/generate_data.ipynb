```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

from utils import *
```

```python
num_points = 2000
num_samples = 10000
sinewave_range = [-3.0, 3.0]
c_range = [-1, 1]

x, c, y = generate_sine_data(num_points, num_samples, sinewave_range, c_range)

print(f'x shape: {x.shape}')
print(f'c shape: {c.shape}')
print(f'y shape: {y.shape}')
```

```python
selected_indices = np.random.choice(num_samples, 6, replace=False)

# Create subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Random Samples of Generated Sine Data')

# Plot each selected sample
for i, idx in enumerate(selected_indices):
    row = i // 3
    col = i % 3
    axs[row, col].plot(x, y[idx])
    axs[row, col].set_title(f'Sample {idx+1}: c1={c[idx,0]:.2f}, c2={c[idx,1]:.2f}, c3={c[idx,2]:.2f}')
    axs[row, col].set_xlim(sinewave_range[0], sinewave_range[1])
    axs[row, col].set_ylim(-3, 3)

plt.tight_layout()
plt.show()
```

```python
c_train, c_test, y_train, y_test = train_test_split(c, y, test_size=0.2, random_state=232)

np.save('./data/x.npy', x)
np.save('./data/c_train.npy', c_train)
np.save('./data/y_train.npy', y_train)
np.save('./data/c_test.npy', c_test)
np.save('./data/y_test.npy', y_test)
```
